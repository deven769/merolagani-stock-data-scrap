```python
import json
import requests
import pandas as pd
import csv

url = 'https://merolagani.com/handlers/webrequesthandler.ashx?type=get_company_graph&symbol=SRBL&dateRange=300'
f = requests.get(url)
_f=f.text
```

```python
json_acceptable_string = _f.replace("'", "\"")
d = json.loads(json_acceptable_string)
```

```python
data = d.get('quotes')
```

```python
open = []
high =[]
low =[]
close = []
volume =[]
rsi =[]
date = []



for i in range(len(data)):
   
    op = data[i].get('open')
    open.append(op)
    
    hi = data[i].get('high')
    high.append(hi)
    
    lo = data[i].get('low')
    low.append(lo)
   
    clo = data[i].get('close')
    close.append(clo)
    
    vol = data[i].get('volume')
    volume.append(vol)
    
    rs = data[i].get('rsi')
    rsi.append(rs)
    
    dt = data[i].get('date')
    date.append(dt)
    
    
    
    
    
```

```python
df = pd.DataFrame({'Date':date,'Open':open,'Close':close,'High':high,'Low':low,'Volume':volume,'RSI':rsi})
```
